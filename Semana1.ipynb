```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Define width and height
W = 100
H = 100

# Create a black image and draw a white diagonal line
im1 = np.zeros((W, H, 3), np.uint8)
for i in range(min(W, H)):
    im1[i, i] = (255, 255, 255)  # White color in BGR
cv2_imshow(im1)

# Create a black image and draw nested colored squares
im2 = np.zeros((W, H, 3), np.uint8)

# Draw the outer red square with a thickness of 5 pixels
cv2.rectangle(im2, (10, 10), (90, 90), (0, 0, 255), 5)

# Draw the middle blue square with a thickness of 5 pixels
cv2.rectangle(im2, (30, 30), (70, 70), (255, 0, 0), 5)

# Draw the inner green square with a thickness of 5 pixels
cv2.rectangle(im2, (45, 45), (55, 55), (0, 255, 0), 5)

cv2_imshow(im2)


```
